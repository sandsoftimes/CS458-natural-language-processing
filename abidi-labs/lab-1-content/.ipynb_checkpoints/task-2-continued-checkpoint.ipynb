# TASK-1

```python
corpora = [text1, text2, text3, text4, text5, text6, text7, text8, text9]
corpus_names = ["text1", "text2", "text3", "text4", "text5", "text6", "text7", "text8", "text9"]

table = []
for corpus, name in zip(corpora, corpus_names):
    corpus_length = len(corpus)
    unique_words = len(set(corpus))
    lexical_richness = unique_words / corpus_length
    table.append([name, corpus_length, unique_words, lexical_richness])

import pandas as pd
columns = ["Corpus Name", "Corpus Length", "Unique Words", "Lexical Richness"]
df = pd.DataFrame(table, columns=columns)
print(df)
```

```python
import math
from nltk.probability import FreqDist

def TF(corpus, token):
    return corpus.count(token) / len(corpus) * 100 if len(corpus) > 0 else 0

def LOGTF(corpus, token):
    return math.log(corpus.count(token) + 1, 10)

def IDF(corpus, token):
    return math.log(len(corpora) / (corpus.count(token) + 1), 10)

tokens = ["monster", "evil", "devil", "the"]  # Add more tokens as needed
results = []

for token in tokens:
    tf_values = [TF(corpus, token) for corpus in corpora]
    logtf_values = [LOGTF(corpus, token) for corpus in corpora]
    idf_values = [IDF(corpus, token) for corpus in corpora]
    results.append([token] + tf_values + logtf_values + idf_values)

columns = ["Token", "Text1 TF", "Text2 TF", "Text3 TF", "Text4 TF", "Text5 TF", "Text6 TF", "Text7 TF", "Text8 TF", "Text9 TF",
           "Text1 LOGTF", "Text2 LOGTF", "Text3 LOGTF", "Text4 LOGTF", "Text5 LOGTF", "Text6 LOGTF", "Text7 LOGTF", "Text8 LOGTF", "Text9 LOGTF",
           "Text1 IDF", "Text2 IDF", "Text3 IDF", "Text4 IDF", "Text5 IDF", "Text6 IDF", "Text7 IDF", "Text8 IDF", "Text9 IDF"]

df = pd.DataFrame(results, columns=columns)
print(df)
```

```python
from nltk.corpus import PlaintextCorpusReader

corpus_root = '/home/spoofy/Downloads/Semester 7/CS458-natural-language-processing/abidi-labs/lab-1-content'

corpus = PlaintextCorpusReader(corpus_root, '.*')

corpus_words = corpus.words()
bigram_list = list(nltk.bigrams(corpus_words))

cfd = nltk.ConditionalFreqDist(bigram_list)

def generate_model(cfdist, word, num):
    generated_words = []
    for i in range(num):
        generated_words.append(word)
        word = cfdist[word].max()
    return generated_words

generated_sentence = generate_model(cfd, 'is', 30)

unique_words_in_corpus = len(set(corpus_words))

number_of_words_in_corpus = unique_words_in_corpus
thirty_word_generated_sentence = ' '.join(generated_sentence)

print("Number of Words in Corpus:", number_of_words_in_corpus)
print("30-word Generated Sentence:", thirty_word_generated_sentence)
```
