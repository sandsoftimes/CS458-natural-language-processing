```python
import math
from nltk import FreqDist

# Function to calculate Term Frequency (TF)
def TF(corpus, token):
    return corpus.count(token) / len(corpus)

# Function to calculate Logarithmic Term Frequency (LOGTF)
def LOGTF(corpus, token):
    return math.log10(corpus.count(token) + 1)

# Function to calculate Inverse Document Frequency (IDF)
def IDF(corpus, token):
    return math.log10(len(corpus) / (1 + sum(1 for doc in corpus if token in doc)))


tokens = ["monster", "evil", "devil", "the", "Common word 1", "Common word 2", "Common word 3"]

# Assuming TF, LOGTF, IDF functions are defined elsewhere

for t in tokens:
    tf = TF(text1, t)
    logtf = LOGTF(text1, t)
    idf = IDF(text1, t)

    # Print values in the desired format
    print(f"Token Name: {t}")
    print(f"TF Value: {tf}")
    print(f"LOGTF Value: {logtf}")
    print(f"IDF Value: {idf}")
    print("\n")
```
