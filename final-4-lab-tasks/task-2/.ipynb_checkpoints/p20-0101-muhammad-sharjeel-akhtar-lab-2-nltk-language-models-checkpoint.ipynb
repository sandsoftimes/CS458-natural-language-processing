# TASK NO 1 Table Answer (With Excel File)

```python
from nltk import FreqDist, bigrams, ngrams
from nltk.collocations import TrigramCollocationFinder, TrigramAssocMeasures

# Assuming you have your three texts: text1, text2, text3

# Create a function to get word frequency distribution and unique words
def get_word_stats(text):
    words = [w.lower() for w in text if w.isalpha()]  # Assuming you want to exclude punctuations
    fdist = FreqDist(words)
    unique_words = sorted(set(words))
    return fdist, unique_words

# Function to get bigrams
def get_bigrams(text):
    return list(bigrams(text))

# Function to get trigrams
def get_trigrams(text):
    return list(ngrams(text, 3))

# Function to get collocations for trigrams
def get_trigram_collocations(text):
    trigram_measures = TrigramAssocMeasures()
    finder = TrigramCollocationFinder.from_words(text)
    return finder.nbest(trigram_measures.pmi, 5)

# Function to get words with length > 16
def get_words_length_gt_16(unique_words):
    return len([w for w in unique_words if len(w) > 16])

# Function to get words with frequency > 500
def get_words_freq_gt_500(fdist):
    return len([w for w in fdist if fdist[w] > 500])

# Function to get words ending in "ed"
def get_words_ending_in_ed(unique_words):
    return len([w for w in unique_words if w.endswith("ed")])

# Applying the functions to each text
fdist1, unique_words1 = get_word_stats(text1)
fdist2, unique_words2 = get_word_stats(text2)
fdist3, unique_words3 = get_word_stats(text3)

bigrams1 = get_bigrams(text1)
bigrams2 = get_bigrams(text2)
bigrams3 = get_bigrams(text3)

trigrams1 = get_trigrams(text1)
trigrams2 = get_trigrams(text2)
trigrams3 = get_trigrams(text3)

collocations1 = get_trigram_collocations(text1)
collocations2 = get_trigram_collocations(text2)
collocations3 = get_trigram_collocations(text3)

# For Text 1
print("Text1:")
print("  Number of words with length > 16:", get_words_length_gt_16(unique_words1))
print("  Number of words with frequency > 500:", get_words_freq_gt_500(fdist1))
print("  Number of words ending in 'ed':", get_words_ending_in_ed(unique_words1))
print("  10 frequently occurring Bigrams:\n", bigrams1[:10])
print("  5 frequently occurring Trigrams:\n", collocations1)

# For Text 2
print("\nText2:")
print("  Number of words with length > 16:", get_words_length_gt_16(unique_words2))
print("  Number of words with frequency > 500:", get_words_freq_gt_500(fdist2))
print("  Number of words ending in 'ed':", get_words_ending_in_ed(unique_words2))
print("  10 frequently occurring Bigrams:\n", bigrams2[:10])
print("  5 frequently occurring Trigrams:\n", collocations2)

# For Text 3
print("\nText3:")
print("  Number of words with length > 16:", get_words_length_gt_16(unique_words3))
print("  Number of words with frequency > 500:", get_words_freq_gt_500(fdist3))
print("  Number of words ending in 'ed':", get_words_ending_in_ed(unique_words3))
print("  10 frequently occurring Bigrams:\n", bigrams3[:10])
print("  5 frequently occurring Trigrams\n:", collocations3)
```

# TASK NO 3: Generating Random Text with Bigrams

```python
cfdist = nltk.ConditionalFreqDist(bigramlist)
```

```python
import random

def generate_model(cfdist, word, num):
    generated_sentence = [word]
    for _ in range(num):
        try:
            # Get all words with their frequencies for the given context
            words_with_freq = list(cfdist[word].items())
            
            # Choose a word randomly based on frequencies
            next_word = random.choices(words_with_freq)[0][0]
            
            generated_sentence.append(next_word)
            word = next_word
        except (ValueError, IndexError):
            # Handle the case where the word is not present in the distribution
            break
    return ' '.join(generated_sentence)
```

```python
generate_model(cfdist, 'Peshawar', 10)
```

```python
# Number of words in the corpus
num_words_in_corpus = len(cfdist)
print("Number of Words in Corpus:", num_words_in_corpus)

# Generate a 30-word sentence
generated_sentence = generate_model(cfdist, 'Peshawar', 30)
print("\n30-word generated sentence:", generated_sentence)
```

# TASK NO 3 Table Answer 

```python
def create_table(header, data, table_width=80):
    table = f"+{'=' * (table_width-2)}+\n"
    table += f"|{header.center(table_width-2)}|\n"
    table += f"+{'=' * (table_width-2)}+\n"
    
    # Split the data into lines with a maximum width
    lines = [data[i:i+(table_width-4)] for i in range(0, len(data), (table_width-4))]
    for line in lines:
        table += f"| {line.ljust(table_width-6)} |\n"
    
    table += f"+{'=' * (table_width-2)}+\n"
    return table

# Number of words in the corpus
num_words_in_corpus = len(cfdist)
num_words_table = create_table("Number of Words in Corpus", str(num_words_in_corpus))

# Generate a 30-word sentence
generated_sentence = generate_model(cfdist, 'Peshawar', 30)
sentence_table = create_table("30-word generated sentence for word Peshawar", generated_sentence)

# Display tables
print(num_words_table)
print(sentence_table)
```
